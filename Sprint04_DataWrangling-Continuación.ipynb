    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.

- `departments.csv`

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
# Utilizo las bases de datos que ya había creado para revisar esta proporción en una nueva base de datos.
# En este caso, genero una nueva base de datos de los productos NO reordenados.

df_norep = df_rep[df_rep["reordered"] == 0].groupby(by = ["product_id", "product_name"]).count().sort_values(by = "reordered", ascending = False)
```

```python
# Combino los dataframes para hacer la comparación.
df_repvsrepsum = df_norep.merge(df_repsum, on = ["product_id", "product_name"], how = "outer", suffixes = ("_no", "_si"))

# Lleno los valores nulos con 0 porque al no haber valores no existen órdenes.
df_repvsrepsum["reordered_no"] = df_repvsrepsum["reordered_no"].fillna(0)
df_repvsrepsum["reordered_si"] = df_repvsrepsum["reordered_si"].fillna(0)
```

```python
# Genero una nueva columna con la proporción de lo que sí se reordenó contra el total de las órdenes.

df_repvsrepsum["% de reorden"] = round(100 * (df_repvsrepsum["reordered_si"]/(df_repvsrepsum["reordered_si"] + df_repvsrepsum["reordered_no"])), 2)
print(df_repvsrepsum.head(20))
df_repvsrepsum.describe()
```

Esto significa que la mayoría de los productos no son reordenados, ya que el promedio es menor y no pueden existir valores fuera del rango de 0% a 100% y se puede afirmar con certeza.
